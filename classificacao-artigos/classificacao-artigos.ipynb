# Classificação de disciplinas.
Esta é uma tentativa de agrupar disciplinas semelhantes, somente usando sua ementa.

Conectando por allegro, para pegar algumas ementas.
O arquivo de credenciais foi retirado.

```python
from franz.openrdf.sail.allegrographserver import AllegroGraphServer
from franz.openrdf.repository.repositoryconnection import RepositoryConnection
from franz.openrdf.model import Literal, URI
import json

# pegando creds
with open("creds.json", "r") as f:
    creds: dict = json.load(f)

server = AllegroGraphServer(
    host=creds.get('host'),
    port=creds.get('port'),
    user=creds.get('user'),
    password=creds.get('password'),
)

conn: RepositoryConnection = server.openFederated(repositories=['teste_ementas2'])
```

```python
iri_disciplina = conn.createURI("https://w3id.org/ccso/ccso#KnowledgeBody")
dicionario_disciplinas = {}

with conn.getStatements(predicate=iri_disciplina) as result:
    for r in result:
        sujeito: URI = r.getSubject()
        objeto: Literal = r.getObject()
        code: str = sujeito.getLocalName()
        ementa: str = objeto.toPython()
        dicionario_disciplinas[code] = ementa
```

Agora, com as disciplinas, é possível começar a fazer a análise das ementas

```python
import stanza
from stanza.models.common.doc import Document, Sentence
stanza.download('pt')       # baixa para portugues
pipeline = stanza.Pipeline('pt')       # neural pipeline para portugues
# doc = pipeline(dicionario_disciplinas['disciplinaADM1356'])

dicionario_filtrado = {}

for i, (code, ementa) in enumerate(dicionario_disciplinas.items()):
    ementa_neural: Document = pipeline(ementa)
    for sentenca in ementa_neural.sentences:
        dicionario_filtrado[code] = ' '.join([words.text for words in sentenca.words])
    # para testes
    if i > 10:
        break

print(dicionario_filtrado)
```
